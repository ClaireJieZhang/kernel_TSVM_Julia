```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
all_data = np.loadtxt("labelled_1600points.data", delimiter=",")
```

```python
race_labeled=all_data[:,3]
```

```python
y=all_data[:,0]
```

```python
minority_positive_numbers_before = np.where(np.logical_and(race_labeled == -1,y == 1))[0]
minority_negative_numbers_before = np.where(np.logical_and(race_labeled == -1,y == -1))[0]
majority_positive_numbers_before = np.where(np.logical_and(race_labeled == 1, y == 1))[0]
majority_negative_numbers_before = np.where(np.logical_and(race_labeled == 1, y == -1))[0]
```

```python
plt.figure(figsize=(20, 10))
plt.subplot(1, 2, 1)
plt.scatter(all_data[minority_positive_numbers_before, 1], all_data[minority_positive_numbers_before, 2], color='navy',
            marker='P', lw=0, label="minority pos labeled", s=40)
plt.scatter(all_data[minority_negative_numbers_before, 1], all_data[minority_negative_numbers_before, 2], color='navy',
            marker='8', lw=0, label='minority neg labeled', s=40)


plt.scatter(all_data[majority_positive_numbers_before, 1], all_data[majority_positive_numbers_before, 2], color='c',
            marker='P', lw=0, label="majority pos labeled", s=40)
plt.scatter(all_data[majority_negative_numbers_before, 1], all_data[majority_negative_numbers_before, 2], color='c',
            marker='8', lw=0, label='majority neg labeled', s=40)



plt.legend(scatterpoints=1, shadow=False, loc='upper right')
plt.title("120 labeled data")
plt.subplots_adjust(left=0.07, bottom=0.07, right=0.93, top=0.92)
plt.savefig('plot_of_skewed_data.svg', format='svg', dpi=1200)
plt.show()
```
